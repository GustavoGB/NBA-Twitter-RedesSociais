```python
import pymysql
import json
from functools import partial
import csv

def run_db_query(connection, query, args=None):
    with connection.cursor() as cursor:
        print('Executando query:')
        cursor.execute(query, args)
        for result in cursor:
            print(result)
            
def return_db_results(connection, query, args=None):#Gambiarra para pegar o id
    with connection.cursor() as cursor:
        cursor.execute(query, args)
        for result in cursor:
            return result #retorna logo o primeiro
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='vlm1998',
    database='redes')

db = partial(run_db_query, connection)

```

```python
with open("signed.json", 'r') as f: #LISTA DAS TRANSFERENCIAS COLETADAS NO SCRAPPING
    signed = json.load(f)
    
with open("all_players.json", 'r') as f: #LISTA DOS JOGADORES COLETADOS POR SCRAPPING
    all_players = json.load(f)
    
all_teams = ["Golden State Warriors", "LA Clippers","Boston Celtics", "Chicago Bulls", "Houston Rockets", "Cleveland Cavaliers", 
                "Dallas Mavericks", "Los Angeles Lakers", "Brooklyn Nets", "Memphis Grizzlies", "Washington Wizards", "Oklahoma City Thunder", 
                "Miami Heat", "Minnesota Timberwolves", "Indiana Pacers", "Denver Nuggets", "Charlotte Hornets", "Philadelphia 76ers", 
                "Portland Trail Blazers", "Boston Celtics", "Milwaukee Bucks", "New Orleans Pelicans", "Toronto Raptors", "Phoenix Suns",
                "New York Knicks", "Detroit Pistons", "Utah Jazz", "Sacramento Kings", "Atlanta Hawks", "Orlando Magic", "San Antonio Spurs"]
```

```python
#PRIMERIO PASSO: COLOCAR TODOS JOGADORES E TODOS OS TIMES EM SUAS TABELAS
#SEGUNDO PASSO: PARA CADA TRANSFERENCIA, PROCURAR O ID DO JOGADOR E DO TIME E COLOCAR A TRANSFERENCIA NA TABELA DE TRANSFERENCIAS
```

```python
#Colocando os jogadores no banco de dados
for player in all_players:
    query = ('INSERT INTO jogador (nome) VALUES (%s)')
    arg = player
    run_db_query(connection, query, arg)
```

```python
#Colocando os times no banco de dados
for team in all_teams:
    query = ('INSERT INTO times (nome) VALUES (%s)')
    arg = team
    run_db_query(connection, query, arg)
```

```python
#Inserindo as transferencias no banco de dados
for key in signed:
    query = ('SELECT id_jogador FROM jogador WHERE nome = %s')
    arg = [key]
    id_jogador = return_db_results(connection, query, arg)[0] #Mais uma gambiarra

    query = ('SELECT id_time FROM times WHERE nome = %s')
    arg = signed[key][0]
    id_time = return_db_results(connection, query, arg)[0] 
        
    query = ('INSERT INTO transferencia (id_jogador, id_time, ano) VALUES (%s,%s,%s)')
    arg = [id_jogador, id_time, signed[key][1]]
    run_db_query(connection, query, arg)
```

```python
dict_tags = {}
with open('tags.txt', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        fim = row["Player"].find("\\")
        row["Player"] = row["Player"][:fim]
        dict_tags[row["Player"]] = row["Twitter"]
        line_count += 1
    print(f'Processed {line_count} lines.')

```

```python
#Inserindo as transferencias no banco de dados
for key in dict_tags:
    try:
        query = ('SELECT id_jogador FROM jogador WHERE nome = %s')
        arg = [key]
        id_jogador = return_db_results(connection, query, arg)[0] #Mais uma gambiarra

        query = ('INSERT INTO jogador_tag (id_jogador, tag) VALUES (%s,%s)')
        arg = [id_jogador, dict_tags[key]]
        run_db_query(connection, query, arg)
    except:
        continue
```
