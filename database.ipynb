```python
import pymysql
import json
from functools import partial
import csv

def run_db_query(connection, query, args=None):
    with connection.cursor() as cursor:
        print('Executando query:')
        cursor.execute(query, args)
        for result in cursor:
            print(result)
            
def return_db_results(connection, query, args=None):#Gambiarra para pegar o id
    with connection.cursor() as cursor:
        cursor.execute(query, args)
        for result in cursor:
            return result #retorna logo o primeiro
connection = pymysql.connect(
    host='localhost',
    user='root',
    password='vlm1998',
    database='redes')

db = partial(run_db_query, connection)

```

```python
#Colocando os jogadores no banco de dados
for player in all_players:
    try:
        query = ('INSERT INTO jogador (nome) VALUES (%s)')
        arg = player
        run_db_query(connection, query, arg)
    except Exception as e:
        print(e)
        
```

```python
#Colocando os times no banco de dados
for team in all_teams:
    try:
        query = ('INSERT INTO times (nome) VALUES (%s)')
        arg = team
        run_db_query(connection, query, arg)
    except Exception as e:
        print(e)
```

```python
#Inserindo as transferencias no banco de dados
for key,listas in signed.items():
    query = ('SELECT id_jogador FROM jogador WHERE nome = %s')
    arg = [key]
    id_jogador = return_db_results(connection, query, arg)[0] #Mais uma gambiarra
    for l in listas: # uma lista de listas
        query = ('SELECT id_time FROM times WHERE nome = %s')
        arg = l[0]
        id_time = return_db_results(connection, query, arg)[0] 
        
        query = ('INSERT INTO transferencia (id_jogador, id_time, ano) VALUES (%s,%s,%s)')
        arg = [id_jogador, id_time, l[1]]
        run_db_query(connection, query, arg)
```

```python
#Inserindo as transferencias no banco de dados
for key,listas in waived.items():
    query = ('SELECT id_jogador FROM jogador WHERE nome = %s')
    arg = [key]
    id_jogador = return_db_results(connection, query, arg)[0] #Mais uma gambiarra
    for l in listas: # uma lista de listas
        query = ('SELECT id_time FROM times WHERE nome = %s')
        arg = l[0]
        id_time = return_db_results(connection, query, arg)[0] 
        
        query = ('INSERT INTO waived (id_jogador, id_time, ano) VALUES (%s,%s,%s)')
        arg = [id_jogador, id_time, l[1]]
        run_db_query(connection, query, arg)
```

```python
dict_tags = {}
with open('tags.txt', mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    line_count = 0
    for row in csv_reader:
        if line_count == 0:
            print(f'Column names are {", ".join(row)}')
            line_count += 1
        fim = row["Player"].find("\\")
        row["Player"] = row["Player"][:fim]
        dict_tags[row["Player"]] = row["Twitter"]
        line_count += 1
    print(f'Processed {line_count} lines.')

```

```python
#Inserindo as transferencias no banco de dados
for key in dict_tags:
    try:
        query = ('SELECT id_jogador FROM jogador WHERE nome = %s')
        arg = [key]
        id_jogador = return_db_results(connection, query, arg)[0] #Mais uma gambiarra

        query = ('INSERT INTO jogador_tag (id_jogador, tag) VALUES (%s,%s)')
        arg = [id_jogador, dict_tags[key]]
        run_db_query(connection, query, arg)
    except:
        continue
```
